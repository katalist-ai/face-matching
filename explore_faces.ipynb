```python
import os
import random
import requests
import pandas as pd
from utils import prepare_number
```

```python
df = pd.read_csv('data/IMDb-Face.csv')
```

```python
L = len(df)
N = 100
random.seed(42)

test_images_dir = 'data/test_images'
if not os.path.exists(test_images_dir):
    os.makedirs(test_images_dir)
```

```python
n = 0
for i in random.sample(range(len(df)), 250):
    url = df.iloc[i]['url']
    if url.endswith('.jpg'):
        response = requests.get(url)
        if response.status_code == 200:
            with open(os.path.join(test_images_dir, prepare_number(n) + '.jpg'), 'wb') as file:
                file.write(response.content)
            n += 1
        else:
            print(f'Failed to download image from {url}')
    else:
        print(f'URL {url} does not end with .jpg')
```
